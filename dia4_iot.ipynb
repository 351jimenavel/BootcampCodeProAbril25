// const hace que no se sobreescriba la variable dada. (Tambien se usa en JavaScript)
/// Se definen 2 pines que no van a variar
const int pinBoton = 1;
const int pinLed = 7;

/// Se definen estados que si pueden variar.
int estadoBoton = 0;
int ultimoEstadoBoton = 0;
bool estadoLed = false;

void setup(){
  pinMode(pinBoton, INPUT_PULLUP); // puede cambiar su estado manteniendolo.
  pinMode(pinLed, OUTPUT);
  digitalWrite(pinLed, LOW); // para que este apagado inicialmente
  Serial.begin(9600)
  Serial.print("Inició el programa, aprieta el boton para encender o apagar");
}

void loop(){
  delay(50);
  estadoBoton = digitalRead(pinBoton); // esto es para leer el estado del pin del boton.
  if (estadoBoton == LOW) {
    estadoLed =! estadoLed;
  digitalWrite(pinLed, estadoLed ? HIGH : LOW); // inicialmente su estado será encendido, y si no, estará apagado
  Serial.print("LED");
  Serial.println(estadoLed ? "ENCENDIDO": "APAGADO");
  ultimoEstadoBoton = estadoBoton;}

}
